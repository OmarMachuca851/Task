# ***Problem:*** *The day when chestnut manju covers the solar system*

```python
import math
import matplotlib.pyplot as plt


CHESNUT_BUN_VOLUME = math.pi * 0.05**2 * 0.01 #volume of a chestnut bun in m³, r: 5cm, h: 1cm
SOLAR_SYSTEM_VOLUME = 4/3 * math.pi * (4.545e12)**3 #volume of the solar system in m³
DOUBLING_TIME_MINUTES = 5 #time taken for the volume to double in minutes

#Function to calculate volume doubling over time
def calculate_doubling_times(volume_1, volume_2):
    doubling_times = 0
    volumes = [volume_1] #initialize volume list with initial volume
    minutes = [0] #initialize minute list with 0
    while volume_1 < volume_2:
        volume_1 *= 2 #double the volume
        doubling_times += 1
        volumes.append(volume_1)
        minutes.append(doubling_times * DOUBLING_TIME_MINUTES)
    return doubling_times, volumes, minutes

#Calculate the time required to fill the solar system whith chestnut buns
doubling_times, volumes, minutes = calculate_doubling_times(CHESNUT_BUN_VOLUME, SOLAR_SYSTEM_VOLUME)
total_time_minutes = minutes[-1]

print(f"It takes {total_time_minutes} minutes to fill the solar system with chestnut buns.")

#plot the results
plt.plot(minutes, volumes, label='Volume of chestnut buns')
plt.axhline(y=SOLAR_SYSTEM_VOLUME, color='r', linestyle='--', label='Solar system volume')
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (m³)")
plt.title("Chestnut buns Volume Doubling Over Time")
plt.legend()
plt.show()
```

```python
def simulated_doubling(initial_volume, target_volume, doubling_time_minutes):
    doubling_times, volumes, minutes = calculate_doubling_times(initial_volume, target_volume)
    total_time_minutes = doubling_times * doubling_time_minutes

    print(f"It will take approximately {total_time_minutes} minutes to reach the target volume")

    #plot the results
    fontdict  = {'family':'serif', 'color':'darkblue', 'weight':'normal', 'size':16}

    plt.plot(minutes, volumes, label='Volume over time')
    plt.axhline(y=target_volume, color='r', linestyle='--', label='Target volume')
    plt.xlabel("Time (minutes)", fontdict=fontdict)
    plt.ylabel("Volume (m³)",fontdict=fontdict)
    plt.title("Volume doubling over time", fontdict=fontdict)
    plt.legend()
    plt.show()


#Example usage for Tokyo Dome
TOKYO_DOME_VOLUME = 1.24e6 #m³
simulated_doubling(CHESNUT_BUN_VOLUME, TOKYO_DOME_VOLUME, DOUBLING_TIME_MINUTES)
```
