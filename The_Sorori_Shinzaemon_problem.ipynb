# ***Problem 1:*** *Number of rice grains on the 100th day*

```python
import matplotlib.pyplot as plt

grain = 1
grain_per_day = []
total_grain_per_day = []

for day in range(100):
  if day == 0:
    grain_per_day.append(grain)
    total_grain_per_day.append(grain)
  else:
    grain *= 2
    grain_per_day.append(grain)
    total_grain_per_day.append(grain + total_grain_per_day[day-1])

print(f'Number of grains of rice on the 100th day: {grain_per_day[-1]}')
print(f'Total number of grains of rice: {total_grain_per_day[-1]}')

fontdict  = {'family':'serif', 'color':'darkblue', 'weight':'normal', 'size':16}

fig, ax = plt.subplots()
ax.plot(grain_per_day, label='Number of grains of rice per day', linestyle='-.')
ax.plot(total_grain_per_day, label='Total number of grains of rice per day', linestyle='--')
ax.set_xlabel('Day', fontdict=fontdict)
ax.set_ylabel('Number of grains of rice', fontdict=fontdict)
ax.set_title('The Sorori Shinzaemon problem', fontdict=fontdict)
ax.legend()
plt.show()
```

# ***Problem 2:*** *Number of rice grains other than the 100th day*

```python
def compute_Sorori_Shinzaemon(n=100): #numero de dias por defecto 100 dias
  grain = 1
  grain_per_day = []
  total_grain_per_day = []

  for day in range(n):
    if day == 0:
      grain_per_day.append(grain)
      total_grain_per_day.append(grain)
    else:
      grain *= 2
      grain_per_day.append(grain)
      total_grain_per_day.append(grain + total_grain_per_day[day-1])

  return grain_per_day, total_grain_per_day

list_grain_per_day, list_total_grain_per_day = compute_Sorori_Shinzaemon(n=31)
print(f'Number of grains of rice on the {len(list_grain_per_day)}th day: {list_grain_per_day[-1]}')
print(f'Total number of grains of rice: {list_total_grain_per_day[-1]}')
```

# ***Problem 3:*** *How many people can survive on the rice they receive for how many days?*

```python
def compute_days_of_rice(total_grains, number_of_people):
  daily_rice_grain_per_person = int(100/0.02) #en promedio una personacondume 100g de arroz,
                                            #un grano de arroz pesa aproximadamente 0.02g
  total_daily_consumption = daily_rice_grain_per_person * number_of_people

  days_of_rice = total_grains // total_daily_consumption

  return days_of_rice

total_no_rice_grains = list_total_grain_per_day[-1]
number_of_people = 30
days_people_can_live = compute_days_of_rice(total_no_rice_grains, number_of_people)
print(f'with {total_no_rice_grains} grains of rice, {number_of_people} people can live for {days_people_can_live} days')
```
