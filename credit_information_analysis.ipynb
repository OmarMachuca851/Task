### Comprobar valores faltantes

```python
# Count missing values
missing  = application_df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

missing_percentage = (missing / len(application_df)) * 100
missing_df = pd.DataFrame({'Missing Vallues': missing, 'Percentage': missing_percentage})
missing_df
```

### dibujar un gráfico que muestre el prorcentaje de clases

```python
import matplotlib.pyplot as plt

#Count instances of each calss
target_counts = application_df['TARGET'].value_counts()

#Create a pie chart
plt.figure(figsize=(10, 6))
plt.pie(target_counts, labels=['Clients who repaid the loan (0)', 'Unpaid loan(1)'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'salmon'])
plt.axis('equal') # Ensures the pie chart if circular
plt.title('distribution of loan repayment status')
plt.legend(loc=1, )
plt.show()
```

El resultado muestra que el 91.9% de los solicitantes reembolsaron sus préstamos, mientras el 8.1% incumplieron. ¿Qué significa esto?

- Esto confirma que losimpagos de prestamos son ralativamente raros.

- Dado que los conjuntos de datos estan desequilibrados, lo que significa que hay muchos menos casos de impagos, los modelos de predicción podrían tener dificultades para aprender patrones de prestamos impagos.

- Un prestamista típico esperaría que los incumplimientos fueran mucho menores que los préstamos reembolsados, lo que se alinea con el comportamiento crediticio del mundo real

## Problema 4: Exploración de datos

### 1. ¿Afectan los niveles de ingresos al impago de préstamos?

```python
import seaborn as sns

sns.kdeplot(data=application_df[application_df['TARGET'] == 0], x='AMT_INCOME_TOTAL', label='No Default')
sns.kdeplot(data=application_df[application_df['TARGET'] == 1], x='AMT_INCOME_TOTAL', label='Default')
plt.title('Income Distribution by Loan Default Status')
plt.legend()
plt.show()
```

### 2. ¿El monto del préstamo o el monto del crédito se relacionan con la probabilidad de incumplimiento?

```python
sns.boxplot(x='TARGET', y='AMT_CREDIT', data=application_df)
plt.title('Loan Credit Amount vs Default')
plt.xticks([0, 1], ['No default', 'Default'])
plt.show()
```

### 3. ¿Son ciertas ocupaciones o tipos de empleos mas riesgosos que otros?

```python
occ_df = application_df[['OCCUPATION_TYPE', 'TARGET']].groupby('OCCUPATION_TYPE').mean().sort_values(by='TARGET', ascending=False)
occ_df.plot(kind='bar', legend=False)
plt.ylabel('Default Rate')
plt.title('Default Rate by Ocupation')
plt.xticks(rotation=90)
plt.show()
```

### 4. ¿La edad juega un papel en la probabilidad de incumplimiento?

```python
application_df['AGE_YEARS'] = application_df['DAYS_BIRTH'] / -365

sns.kdeplot(data=application_df[application_df['TARGET'] == 0], x='AGE_YEARS', label='no Default')
sns.kdeplot(data=application_df[application_df['TARGET'] == 1], x='AGE_YEARS', label='Default')
plt.title('Age distribution by Loan Default status')
plt.legend()
plt.show()
```

### 5. ¿El estado familiar (número de niños o estado civil) estan vinculado a las tarifas predeterminadas?

```python
child_df = application_df.groupby('CNT_CHILDREN')['TARGET'].mean()

child_df.plot(kind='bar')
plt.ylabel('Default Rate')
plt.title('Default Rate by Number of Children')
plt.xlabel('number of children')
plt.show()
```
