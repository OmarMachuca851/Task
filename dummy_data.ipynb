## ***Problem 1: Creating random numbers***

```python
import numpy as np

mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]

np.random.seed(0)

random_number = np.random.multivariate_normal(mean, cov, 500)

print(random_number)
random_number.shape
```

## ***Problem 2: Visualization by scatter plot***

```python
import matplotlib.pyplot as plt
import numpy as np

mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]

np.random.seed(0)

x, y = np.random.multivariate_normal(mean, cov, 500).T

plt.title('scatter')
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(x, y)
plt.axis('equal')
plt.show()
```

## ***Problem 3: Visualization by histogram***

```python
plt.xlabel('x1')
plt.ylabel('frecuancy')
plt.title('histogram of x1')

x, y = np.random.multivariate_normal(mean, cov, 500).T
plt.hist(x, color='grey')
plt.xlim(right=3)
plt.show()

plt.xlabel('x2')
plt.ylabel('frecuancy')
plt.title('histogram of x2')

plt.hist(y)
plt.xlim(left=-8)
plt.show()
```

## ***Problem 4: Additon of Data***

```python
mean2 = (0, -3)
cov2 = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)

x, y = np.random.multivariate_normal(mean, cov, 500).T
x2, y2 = np.random.multivariate_normal(mean2, cov2, 500).T

plt.title('scatter')
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(x, y, alpha=0.5, label='0')
plt.scatter(x2, y2, alpha=0.5, color='red', label='1')
plt.legend(loc='upper right')

plt.show()
```

## ***Problem 5: Data combination***

```python
random_number1 = np.random.multivariate_normal(mean, cov, 500)
random_number2 = np.random.multivariate_normal(mean2, cov2, 500)

np.random.seed(0)

random_numbers_combined = np.vstack((random_number1, random_number2))

print(random_numbers_combined)
random_numbers_combined.shape
```

## ***Problem 6: Labelling***

```python
zeros_ndarray = np.zeros(500)
ones_ndarray = np.ones(500)

first_ndarray = np.insert(random_number1, 2, zeros_ndarray, axis=1)
second_ndarray = np.insert(random_number2, 2, ones_ndarray, axis=1)
np.random.seed(0)
labeled_ndarray = np.concatenate((first_ndarray, second_ndarray))

print(labeled_ndarray)
labeled_ndarray.shape
```
