# ***Two-dimensional array and gradient problem***

## ***Problem 1: Linear Function***

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-50, 50.1, 0.1) # 50.1 to inckude 50

# Calculate y = x/2 +1
y = 1/2 * x + 1
print(f'element ef x: {x}')
print(f'element of y: {y}')
```

## ***Problem 2: Array combination***

```python
xy = np.column_stack((x,y))

print(f'shapes of combined array: {xy.shape}')
print(f'Combined array elements: {xy}')
```

## ***Problem 3: Find the gradient***

```python
dy = np.diff(y)
dx = np.diff(x)
gradient = dy/dx

print(f'gradient values: {gradient}')
print(f'lenght of gradient array: {len(gradient)}')
```

## ***Problem 4: Draw a graph***

```python
# x for the gradient has one less value, so we trim x
x_gradient = x[:-1]

# set up the plots
plt.figure(figsize=(12, 5))

# plot for the original function (linear function)
plt.subplot(1, 2, 1)
plt.plot(x, y, label='y = 0.5x +1', color='blue')
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# plot the gradient
plt.subplot(1, 2, 2)
plt.plot(x_gradient, gradient, label='gradient', color='green')
plt.title('Gradient at Each x')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.grid(True)
plt.legend()

#show both plots
plt.tight_layout()
plt.show()
```

## ***Problem 5: python functionalization***

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(*x_range)
    y = function(x)
    dy = np.diff(y)
    dx = np.diff(x)

    array_xy = np.column_stack((x, y))
    gradient = dy/dx
    return array_xy, gradient

#compute required function
def function1(x):
    return x**2

def function2(x):
    return 2*x**2 + 2*x

def function3(x):
    return np.sin(x/12)

xy1, grad1 = compute_gradient(function1)

xy2, grad2 = compute_gradient(function2)

xy3, grad3 = compute_gradient(function3, x_range=(0, 100.1, 0.1))
```

```python
def function_gradient_plot(xy, gradient, tile_prefix):
    x = xy[:, 0]
    y = xy[:, 1]
    x_grad = x[:-1]

    plt.figure(figsize=(12,5))

    # plot for the original function (linear function)
    plt.subplot(1, 2, 1)
    plt.plot(x, y, label=f'{tile_prefix} Function')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{tile_prefix} Function')
    plt.grid(True)
    plt.legend()

    # plot the gradient
    plt.subplot(1, 2, 2)
    plt.plot(x_grad, gradient, label=f'{tile_prefix} Gradient', color='orange')
    plt.xlabel('x')
    plt.ylabel('gradient')
    plt.title(f'{tile_prefix} Gradient')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

function_gradient_plot(xy1, grad1, 'y = x')
function_gradient_plot(xy2, grad2, 'y = 2x + 2x')
function_gradient_plot(xy3, grad3, 'y = sin(x/12)')
```

## ***Problem 6: Finding the minimum value***

```python
def analyze_minimum(xy, gradient, func_label):
    y = xy[:,1]
    x = xy[:,0]

    min_index = np.argmin(y)
    min_y = y[min_index]
    min_x = x[min_index]

    print(f'\n---{func_label}---')
    print(f'minimun y = {min_y} at x = {min_x:.4f}')

    #show slope before and after, if they exist
    if 1 <= min_index <len(gradient):
        print(f'slope before minimum: {gradient[min_index-1]:.4f}')
        print(f'slope after minimum: {gradient[min_index]:.4f}')
    else:
        print('minimun is at the edge of domain, cannot compute both slopes.')


xy1, grad1 = compute_gradient(function1)
analyze_minimum(xy1, grad1, 'y = x')

xy2, grad2 = compute_gradient(function2)
analyze_minimum(xy2, grad2, 'y = 2x + 2x')

xy3, grad3 = compute_gradient(function3, (0, 100.1, 0.1))
analyze_minimum(xy3, grad3, 'y = sen(x/12)')
```
