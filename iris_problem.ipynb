## ***Problem 1: data Acquisition***

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

#create DataFrame X with feature values
X = pd.DataFrame(
    data=iris.data,
    columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
)

#Create DataFrame y with target values (species)
y = pd.DataFrame(data=iris.target, columns=['Species'])

X
```

## ***Problem 2: Combining data***

```python
df = pd.concat([X, y], axis=1)
df
```

### Difference between .loc and .iloc:
- `.loc`: Label-based indexing. Uses the actual row/column labels to select data.
- `.iloc`: Integer position-based indexing. Uses numerical indicies (0, 1, 2, 3,.....) to select data.

Example:
- `df.loc[10, 'sepal_width']` selects the vlue at row 10, column 'sepal_width'
- `df.iloc[10, 1]` selects the value at the 1th row, 2nd column

## ***Problem 6: Creating a diagram***

### Pie chart of species distribution

```python
import matplotlib.pyplot as plt

# Convet numerical species to actual names
species_names = {0: 'Iris setosa', 1: 'Iris versicolor', 2: 'Iris virginica'}
df['species_names'] = df['Species'].map(species_names)

#Pie chart sample counts per label
plt.figure(figsize=(10, 6))
species_counts = df['Species'].value_counts()
plt.pie(species_counts, labels=[species_names[i] for i in species_counts.index],
        autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Distribution of Iris Spaecies')
```

### Box plot each features by species

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='Species', y =feature, data=df)
    plt.title(f'{feature.title()} Distribution')

plt.tight_layout()
plt.show()
```

### Violin Plots feature distribution by species

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.violinplot(x='Species', y =feature, data=df, inner='quartile')
    plt.title(f'{feature.title()} Distribution')

plt.tight_layout()
plt.show()
```

Differences between box plot and violin plot:
- **Box plot**: Displays the median, quartiles, and potential outliers. It provides a summary of the data distribution.
- **Violin plot**: Combines box plot features with a density plot, showing the distribution shape of the data. It provides more information about the data distribution

### Create scatter plots of all combinations (6) such that 2 out of 4 features are selected by color:

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10))

fig.suptitle('- Create scatter plots of all combinations (6) such that 2 out of 4 features are slected by color.')

sns.scatterplot(ax=ax[0,0], data=df, x='sepal_length', y='sepal_width', hue='Species', palette='deep')
sns.scatterplot(ax=ax[0,1], data=df, x='sepal_length', y='petal_length', hue='Species', palette='deep')
sns.scatterplot(ax=ax[0,2], data=df, x='sepal_length', y='petal_width', hue='Species', palette='deep')
sns.scatterplot(ax=ax[1,0], data=df, x='sepal_width', y='petal_length', hue='Species', palette='deep')
sns.scatterplot(ax=ax[1,1], data=df, x='sepal_width', y='petal_width', hue='Species', palette='deep')
sns.scatterplot(ax=ax[1,2], data=df, x='petal_length', y='petal_width', hue='Species', palette='deep')
```
